# CREATING A FUCTION TO DETECT DEFECT 

 Here we are comparing our sample images with an ideal image, whenever a element mismatches we know that our sample image is 
 defective and it returns its index for getting the location.
 The basic logic is that whenever a 1 is replaced by 0 i.e black on white we have a flash, simillarly whenever a 0 is replaced
 by a 1 i.e white on black we have a cut.
 

```python
def DefectDetect(good_img,sample_img):
    
    if(len(good_img) <= len(sample_img)):
        a=len(good_img) 
    else:
        a=len(sample_img)                                             #running loop for the image with shorter length 

    for i in range(a):
        if(len(good_img[i]) <= len(sample_img[i])):
            b=len(good_img[i])
        else:
            b=len(sample_img[i])                                     #since its a 2D array running loop for shorter inner array
        check=False
        for j in range(b):

            if(good_img[i][j]!=sample_img[i][j]):
                if(sample_img[i][j]==1):                            #comparing each element 
                    print("This is a defective image with a cut")  
                    print('index is [',i,']','[',j,']')
                else:
                    print('This is a defective image with a flash')
                    print('index is [',i,']','[',j,']')
                check=True
               
            else:
                continue 
        if(check==True):
            break
    
```

# IMAGE 1 

converting each image to a numpy array and the passing it as a parameter to our fuction DefectDetect()

```python
A = Image.open('defect1.png').convert('L')

img1 = np.array(A)

img1[img1 > 0] = 1
```

```python
DefectDetect(good_img,img1)
plt.imshow(img1)
```
